```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_data = pd.read_csv('spotify_dataset.csv')
```

```python
spotify_data.drop(['Song ID','Index'], axis = 1 , inplace= True)
```

```python
spotify_data.head()
```

```python
spotify_data.columns
```

```python
for col in spotify_data.columns:
    if pd.api.types.is_string_dtype(spotify_data[col]):
        spotify_data[col] = spotify_data[col].str.strip()
spotify_data = spotify_data.replace({"":np.nan})
```

```python
spotify_data.dropna(inplace=True)
```

```python
spotify_data.reset_index(inplace=True)
```

```python
for col in ['Acousticness','Energy','Danceability','Tempo','Liveness','Speechiness','Valence','Loudness']:
    spotify_data[col] = spotify_data[col].astype(float)
for col in ['Artist Followers','Duration (ms)','Number of Times Charted']:
    spotify_data[[col]].apply(pd.to_numeric)
spotify_data['Streams'] = spotify_data['Streams'].str.replace(',', '').apply(pd.to_numeric)
spotify_data['Duration (ms)'] = spotify_data['Duration (ms)'].astype(int)
spotify_data['Popularity'] = spotify_data['Popularity'].astype(int)
spotify_data['Artist Followers'] = spotify_data['Artist Followers'].astype(int)
```

## Categorical EDA

```python
all_genres = dict()
for i in range(len(spotify_data['Genre'])):
    intermediate =spotify_data['Genre'][i].split("'") 
    if '[' in intermediate[:]:
        intermediate.remove('[')
    if ']' in intermediate[:]:
        intermediate.remove(']')
    if '[]' in intermediate[:]:
        intermediate.remove('[]')
    if ', ' in intermediate[:]:
        intermediate = list(filter(lambda a : a != ', ',intermediate))
    for word in intermediate:
        if word in all_genres:
            all_genres[word] = all_genres[word] + 1
        else:
            all_genres[word] = 1
```

```python
from itertools import islice

def take(n, my_dict):
    """Return the first n items of the iterable as a list."""
    return list(islice(my_dict, n))
```

```python
sorted_tuples =sorted(all_genres.items(), key=lambda x:x[1],reverse=True) 
sorted_dict = dict(sorted_tuples)
top20_genres = take(20, sorted_dict)
sorted_data = []
for i in top20_genres:
    sorted_data.append(all_genres[i])
plt.bar(top20_genres, sorted_data)
plt.xticks(rotation = 'vertical')
```

```python
genre_names = take(1545, all_genres)
streams_on_genre = dict()
for genre in genre_names:
    streams = 0
    for index, row in spotify_data.iterrows():
        if genre in row['Genre']:
            streams += row['Streams']
    streams_on_genre[genre] = streams
```

```python
sorted_tuples2 =sorted(streams_on_genre.items(), key=lambda x:x[1],reverse=True) 
sorted_dict2 = dict(sorted_tuples2)
top20_genres2 = take(20, sorted_dict2)
sorted_data2 = []
for i in top20_genres2:
    sorted_data2.append(streams_on_genre[i])
plt.bar(top20_genres2, list(map(lambda x: x//(10**8), sorted_data2)))
plt.xticks(rotation = 'vertical')
```

```python
def map_chords(x):
    if x == "G#/Ab":
        return 1
    elif x == "A":
        return 2
    elif x == "D#/Eb":
        return 3
    elif x == "F#/Gb":
        return 4
    elif x == "E":
        return 5
    elif x == "F":
        return 6
    elif x == "B":
        return 7
    elif x == "C#/Db":
        return 8
    elif x == "A#/Bb":
        return 9
    elif x == "G":
        return 10
    elif x == "D":
        return 11
    else:
        return 12
```

```python
spotify_data['Chord'] = spotify_data['Chord'].map(lambda x : map_chords(x))
```

#### Question # : Is there a relation between the artists with most streams and artists with highest number of chartings?

```python
artist_times_charted = spotify_data.groupby('Artist')['Number of Times Charted'].sum().sort_values(ascending= False).to_frame().reset_index()
```

```python
artist_times_charted.info()
```

```python
plt.figure()
sns.barplot(x = 'Number of Times Charted', y = 'Artist', data = artist_times_charted[0:20])
plt.title('Top 20 Chartings vs artists')
plt.xlabel('# of times charted')
plt.ylabel('Artist Name')
```

```python
songs_times_charted = spotify_data.groupby('Song Name')['Number of Times Charted'].sum().sort_values(ascending= False).to_frame().reset_index()
```

```python
plt.figure()
sns.barplot(x = 'Number of Times Charted', y = 'Song Name', data = songs_times_charted[0:20])
plt.title('Top 20 Chartings vs songs')
plt.xlabel('# of times charted')
plt.ylabel('Song Name')
```

```python
artist_number_of_songs = spotify_data['Artist'].value_counts().sort_values(ascending = False).reset_index()
```

```python
artist_number_of_songs.info()
```

```python
plt.figure()
sns.barplot(x = 'Artist', y = 'index', data = artist_number_of_songs[0:20])
plt.title('Top 20 artists in terms of # of songs')
plt.xlabel('# of songs')
plt.ylabel('Artist')
```
